```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters of the magnetic field
B = np.array([0, 0, 1])  # Magnetic field vector
q = 1.60217662e-19  # Charge of the particle
m = 9.10938356e-31  # Mass of the particle
v0 = np.array([1e6, 0, 0])  # Initial velocity of the particle
r0 = np.array([0, 0, 0], dtype=np.float64)  # Initial position of the particle

# Define the simulation parameters
tmax = 1e-7  # Maximum simulation time
dt = 1e-10  # Time step

# Initialize the particle position and velocity
r = r0
v = v0

# Initialize lists to store the particle positions and times
x = [r[0]]
y = [r[1]]
z = [r[2]]
t = [0]

# Perform the simulation
while t[-1] < tmax:
    # Calculate the Lorentz force on the particle
    F = q * np.cross(v, B)
    
    # Calculate the acceleration of the particle
    a = F / m
    
    # Update the particle velocity and position
    v += a * dt
    r += v * dt
    
    # Append the particle position and time to the lists
    x.append(r[0])
    y.append(r[1])
    z.append(r[2])
    t.append(t[-1] + dt)

# Plot the magnetic field lines
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Charged Particle Motion in a Uniform Magnetic Field', fontsize=16)



plt.show()
```
